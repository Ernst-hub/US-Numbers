## **Data Preparation**

```python
%%capture
start_time = datetime.datetime(1990,1,1)

### Recessions

data_series4 = 'USREC'
recession = pdr.DataReader(data_series4, 'fred', start_time)
recession.index = pd.to_datetime(recession.index)
recession["Date"] = recession.index
recs = recession.query('USREC==1')
rec_date = recs["Date"]

### Get Unemployment Data: https://fred.stlouisfed.org/series/UNRATE ###

data_series = 'UNRATE'
unrate = pdr.DataReader(data_series, 'fred', start_time)
#drop NAs
unrate = unrate.dropna()
# data frame index to datetime object
unrate.index = pd.to_datetime(unrate.index)
# append date to data frame
unrate["Date"] = unrate.index


### Get All Employess, Total Nonfarm (seasonally adjusted) Data: https://fred.stlouisfed.org/series/PAYEMS ###

data_series2 = 'PAYEMS'
payems = pdr.DataReader(data_series2, 'fred', start_time)
#drop NAs
payems = payems.dropna()
# data frame index to datetime object
payems.index = pd.to_datetime(payems.index)
# append date to data frame
payems["Date"] = payems.index

### average hourly earnings

data_series3 = 'CES0500000003'
avgwage = pdr.DataReader(data_series3, 'fred', start_time)
avgwage.index = pd.to_datetime(avgwage.index)
avgwage["Date"] = avgwage.index

### Labor Force Participation

lfpr_data = 'CIVPART'
lfpr = pdr.DataReader(lfpr_data, 'fred', start_time)
lfpr.index = pd.to_datetime(lfpr.index)
lfpr["Date"] = lfpr.index

### CPI

inflation = 'CPIAUCSL'
INF = pdr.DataReader(inflation, 'fred', start_time)
#drop NAs
INF = INF.dropna()
# data frame index to datetime object
INF.index = pd.to_datetime(INF.index)
# append date to data frame
INF["Date"] = INF.index

#### Core CPI
core_inflation = 'CPILFESL'
CINF = pdr.DataReader(core_inflation, 'fred', start_time)
CINF = CINF.dropna()
CINF.index=pd.to_datetime(CINF.index)
CINF["Date"] = CINF.index

### GDP
#### real gdp
rgdp = 'GDPC1'
GDP = pdr.DataReader(rgdp, 'fred', start_time)
GDP = GDP.dropna()
GDP.index = pd.to_datetime(GDP.index)
GDP["Date"] = GDP.index

```

#### **Inflation**

**Calculate rate percentage change from year ago on**: Consumer Price Index for All Urban Consumers: All Items in U.S. City Average  

```python
i = 0
a = len(INF)-12
DFPCY = []
#load percentages into list
while i < a:
    d = (INF.iloc[(i+12),0]-INF.iloc[i,0])/INF.iloc[i,0]*100
    DFPCY.append(d)
    i+=1

INF2 = INF.iloc[12:,:]
INF2["PCY"] = DFPCY

#test that list and df are the same size

if len(INF2) - len(DFPCY) == 0: 
    print("same size")
else:
    print("not same size")
```

**Do the same for core inflation**

```python
i = 0
a = len(CINF)-12
CINFPCY = []
#load percentages into list
while i < a:
    x = (CINF.iloc[(i+12),0]-CINF.iloc[i,0])/CINF.iloc[i,0]*100
    CINFPCY.append(x)
    i+=1

CINF2 = CINF.iloc[12:,:]
CINF2["CINFPCY"] = CINFPCY

#test that list and df are the same size

if len(CINF2) - len(CINFPCY) == 0: 
    print("same size")
else:
    print("not same size")
```

**YOY Change in GDP**

```python
i = 0
a = len(GDP)-12
DFGDP = []
#load percentages into list
while i < a:
    x = (GDP.iloc[(i+12),0]-GDP.iloc[i,0])/GDP.iloc[i,0]*100
    DFGDP.append(x)
    i+=1

GDP2 = GDP.iloc[12:,:]
GDP2["GDPYOYPC"] = DFGDP

#test that list and df are the same size

if len(GDP2) - len(DFGDP) == 0: 
    print("same size")
else:
    print("not same size")
```
